```python
# STEP 1: Clone YOLOv5 repo and install dependencies
!git clone https://github.com/ultralytics/yolov5
%cd yolov5
!pip install -r requirements.txt

# STEP 2: Import libraries
import torch
from IPython.display import Image, display
import os

# STEP 3: Load a pretrained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # small model, fast

# STEP 4: Upload an image
from google.colab import files
uploaded = files.upload()

# STEP 5: Perform object detection on the uploaded image
import PIL
for filename in uploaded.keys():
    results = model(filename)
    results.save()  # saves to 'runs/detect/exp'
    display(Image(filename=f'runs/detect/exp/{filename}'))
```
